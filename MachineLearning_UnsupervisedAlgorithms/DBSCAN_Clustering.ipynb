```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
```

## make_moons method:
### to generate a synthetic dataset that resembles two interleaved half-circles or "moons." It is often used for testing and demonstrating the performance of clustering and classification algorithms

```python
X,y=make_moons(n_samples=250,noise=0.05)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
##feature scaling(Standard Scaling)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
dbscan=DBSCAN(eps=0.3)
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=dbscan.labels_)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```
